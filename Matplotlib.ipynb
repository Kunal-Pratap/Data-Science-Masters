## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

```python
## Scatter plot plots all the coordinates with dots. Which looks like scattered dots.
```

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.title("X vs Y Graph")
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

```python
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
a = np.array([0, 1, 2, 3, 4, 5])
b = np.array([50, 20, 40, 20, 60, 70])
c = np.array([0, 1, 2, 3, 4, 5])
d = np.array([10, 20, 30, 40, 50, 60])
e = np.array([0, 1, 2, 3, 4, 5])
f = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.plot(x,y, label = "line 1")
plt.plot(a,b, label = "line 2")
plt.plot(c,d, label = "line 3")
plt.plot(e,f, label = "line 4")
plt.legend()
plt.show()
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
## It is a plot which plots the graph as a bar graph.
## It is used to compare between different outcomes.
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
```

```python
plt.barh(company,profit)
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
## Turn the axes box on or off on the current axes.
## It is used to plot box graph.
```
